```python
def change_maker(total, coins, level=0, coin_list=None):
    if total == 0:
        print coin_list
        return
    if coin_list is None:
        coin_list = [0]*len(coins)
    for index in range(level, len(coins)):
#         print total, coins[index]
        if coins[index] <= total:
            coin_list[index] += 1
            change_maker(total-coins[index], coins, index, coin_list)
            coin_list[index] -= 1
```

```python
change_maker(78,[25,10,5,1])
```
