```python
def bubble_sort(arr):
    for i in range(len(arr)-1):
        for j in range(len(arr)- i - 1):
            if arr[j] > arr[j+1]:
                tmp = arr[j+1]
                arr[j+1] = arr[j]
                arr[j] = tmp
                
```

```python
test = [54,26,93,17,77,31,44,55,20]
bubble_sort(test)
```
